```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
# Load Dataset
vehicle_df = pd.read_csv("car_prices.csv")
vehicle_df.shape
```

```python
# will work with data duplicate
df = vehicle_df.copy()
```

```python
# checking for null values in the transaction dataset
df.isnull().sum()
```

```python
# drop null values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# view duplicates
df.duplicated().sum()
```

```python
df.shape
```

```python
df.head(2)
```

```python
# split saledate and achieve the following columns
df['sale_Day'] = df['saledate'].str.split(' ', expand = True)[0]
df['sale_Month'] = df['saledate'].str.split(' ', expand = True)[1]
df['sale_Date'] = df['saledate'].str.split(' ', expand = True)[2].astype(int)
df['sale_year'] = df['saledate'].str.split(' ', expand = True)[3].astype(int)
```

```python
df.head()
```

```python
# drop columns unnessary for our analysis
df = df.drop(columns=['saledate','vin','state'])
```

```python
df
```

```python
# convert column body to lower case
df['body'] = df['body'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
# save dataframe to CSV so we can use in SQL for further analysis
df.to_csv('vehicle-sales.csv')
```
